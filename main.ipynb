# Recommender system implementation using CNN

This is an implementation of the model described in [Collaborative Filtering Recommendation Algorithm Based on Convolutional Neural Network](https://doi.org/10.48550/arXiv.1809.07426).

## Datasets used
Used in the original paper:
1.  [MovieLens 1m](https://grouplens.org/datasets/movielens/1m/) - a data set of 1 million ratings from 6000 users on 4000 movies released 2/2003.
2. [Gowalla](https://snap.stanford.edu/data/loc-gowalla.html) - a location-based social networking website where users share their locations by checking-in. The friendship network is undirected and was collected using their public API, and consists of 196,591 nodes and 950,327 edges. They collected a total of 6,442,890 check-ins of these users over the period of Feb. 2009 - Oct. 2010.
3. [Foursquare]() - user geographical check-in data used in [Graph-based Point-of-interest Recommendation with
Geographical and Temporal Influences](https://personal.ntu.edu.sg/axsun/paper/cikm14-yuan.pdf). I couldn't find the original dataset so I used the ones found in [Dingqi Yang's homepage](https://sites.google.com/site/yangdingqi/home/foursquare-dataset) - data is hosted on google drive.
4. [Tmall](https://tianchi.aliyun.com/dataset/42) - user purchase data from a B2C platform in china. Data is hosted on tianchi (alibaba kaggle).
5. [Amazon](https://cseweb.ucsd.edu/~jmcauley/datasets.html#amazon_reviews) - this data comprises of user reviews of products. I use the 2023 version which is downloaded using [huggingface](https://github.com/huggingface/datasets).
